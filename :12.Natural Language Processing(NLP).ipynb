### 6.Movie reviews dataset

1.find documents=[words,category]

2.clean the document

3.find most common and build dict function

4.build training dataset using dict

5.train the model and predict


```python
from nltk.corpus import movie_reviews
```

```python
nltk.download('movie_reviews')
```

```python
movie_reviews.categories()
```

```python
len(movie_reviews.fileids())
```

```python
movie_reviews.fileids('neg')[:10]#only negative reviews
```

```python
movie_reviews.words(movie_reviews.fileids()[5])
```

```python
documents=[]
for category in movie_reviews.categories():
    for fileid in movie_reviews.fileids(category):
        documents.append((movie_reviews.words(fileid),category))
documents[:5]
```

```python
import random
random.shuffle(documents)#randomly select
documents[:5]
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
```

```python
from nltk.corpus import wordnet
def get_simple_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else: 
        return wordnet.NOUN
```

```python
def clean_review(words):
    output_words=[]
    for w in words:
        if w.lower() not in stop:
            pos=pos_tag([w])
            clean_word=lemmatizer.lemmatize(w,pos=get_simple_pos(pos[0][1]))
            output_words.append(clean_word.lower())
    return output_words    
```

```python
document=[(clean_review(document),category) for document,category in documents]
```

```python
document[0]
```

```python
training_documents=document[:1500]
testing_documents=document[1500:]
```

```python
#build feature set
all_words=[]
for doc in training_documents:
    all_words+=doc[0]
```

```python
import nltk
```

```python
freq=nltk.FreqDist(all_words)
common=freq.most_common(3000)
feature=[i[0] for i in common]
feature
```

```python
def get_feature_dict(words):
    current_feature={}
    words_set=set(words)
    for w in feature:
        current_feature[w]=w in words_set
    return  current_feature
    
```

```python
training_data=[(get_feature_dict(doc),category) for doc,category in training_documents]
testing_data=[(get_feature_dict(doc),category) for doc,category in testing_documents]
```

```python
training_data[0]
```

**NLTK classifiers require data to be in this format: array of tuples,where each tuple has dictionary of features and category**

```python
from nltk import NaiveBayesClassifier
```

```python
cl=NaiveBayesClassifier.train(training_data)
```

```python
nltk.classify.accuracy(cl,testing_data)
```

```python
cl.show_most_informative_features(15)
```

**ngram**

example: if not good often show up together, they will be treated as a combination and represent the negative review

ngram-->(2,3)-->min,max of gram

```python
count_vec = CountVectorizer(max_features = 2000,ngram_range=(1,2))
x_train_features = count_vec.fit_transform(x_train)
x_train_features.todense()
```

```python
count_vec.get_feature_names()
```

```python
x_test_features = count_vec.transform(x_test)
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC()
svc.fit(x_train_features, y_train)
```

```python
svc.score(x_test_features, y_test)
```

we can see we get a better result

### TF-IDF

TF: term frequency-->number of term t occurs in document d

IDF: inverse document frequency-->1/document frequency

     document frequency: for a word, df(w)=no.of documents that contains w

tf*idf:we can use Tfidf Vectorizer instead of Countvectorizer

max_df:if a word in all the documents,the frequency is more than 0.8, I am not interested in this word

min_df:if a word in all the documents,the frequency is less than 0.1, I am not interested in this word
